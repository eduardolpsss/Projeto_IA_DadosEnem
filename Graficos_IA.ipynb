```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd 

# Importando planilha do Excel
planilha2020 = pd.read_csv("./drive/MyDrive/trabalho_ia/data/MICRODADOS_ENEM_2020_DF.csv", sep=';' , encoding='latin-1')
planilha2021 = pd.read_csv("./drive/MyDrive/trabalho_ia/data/MICRODADOS_ENEM_2021_DF.csv", sep=';' , encoding='latin-1')

#---------------------------------- Grafico 1 - Total alunos por ano ----------------------------------

# Printar total de participantes por ano
print('Numero total de participantes na planilha Microdados Enem 2020:', planilha2020.NU_PARTICIPANTES.max())
print('Numero total de participantes na planilha Microdados Enem 2021:', planilha2021.NU_PARTICIPANTES.max(), '\n')

NU_PARTICIPANTES_2020 = planilha2020.NU_PARTICIPANTES.max()
NU_PARTICIPANTES_2021 = planilha2021.NU_PARTICIPANTES.max()

# Printando grafico de barras horizontal
x = ['Planilha 2020', 'Planilha 2021']
y = [116932, 67501]

total_NU_PARTICIPANTES_2020_2021 = NU_PARTICIPANTES_2020, NU_PARTICIPANTES_2021
plt.style.use('fivethirtyeight')

for index, value in enumerate(y):
  plt.text(value, index,str(value))

plt.rcParams['figure.figsize'] = [5, 5]

plt.barh(x, total_NU_PARTICIPANTES_2020_2021, align='center', color=['blue', 'red'], edgecolor='black', linewidth="1") 
plt.xlabel("Total alunos por ano")
plt.ylabel("Planilhas")
```
